```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.io.json import json_normalize
from json import dumps
from math import isnan
import json

pd.set_option('display.max_colwidth', 85)
```

```python
df = pd.read_json('episodes.json')
df = df['episodes']
df = json_normalize(df)
```

```python
for i in range(60):
    for sc in df['scenes'].iloc[i]:
        for ch in sc['characters']:
            if ch['name']=='High Septon':
                print(i)
```

```python
episodes = []
for i, ep in df.iterrows():
    season = ep['seasonNum']
    season = 'S' + str(season)
    episode = ep['episodeNum']
    episode = 'E' + str(episode).rjust(2, '0')
    episodes.append(season + episode)
```

```python
personnages = {}
for ep in df['scenes']:
    for sc in ep:
        characters = sc['characters']
        for ch in characters:
            if ch['name'] in personnages:
                personnages[ch['name']] += 1
            else:
                personnages[ch['name']] = 1

personnages = pd.DataFrame(personnages, index=['appearances']).T
personnages = personnages[personnages['appearances'] > 10]
```

```python
personnages = ["Jon Snow", "Bran Stark", "Robb Stark", "Eddard Stark", "Catelyn Stark", "Sansa Stark",
               "Arya Stark", "Theon Greyjoy", "Cersei Lannister", "Jaime Lannister", "Joffrey Baratheon",
              "Sandor Clegane", "Robert Baratheon", "Tommen Baratheon", "Tyrion Lannister", "Daenerys Targaryen",
              "Viserys Targaryen", "Khal Drogo", "The Night King", "High Sparrow", "Euron Greyjoy", "Lord Varys",
              "Renly Baratheon", "Petyr Baelish", "Samwell Tarly", "Gendry", "Gregor Clegane", "Bronn",
               "Loras Tyrell", "Tywin Lannister", "Walder Frey", "Shae", "Gilly", "Melisandre", "Davos Seaworth",
              "Stannis Baratheon", "Podrick Payne", "Yara Greyjoy", "Margaery Tyrell",
               "Brienne of Tarth", "Ygritte", "Tormund Giantsbane", "Qyburn", "Missandei", "Olenna Tyrell",
              "Ramsay Snow", "Grey Worm", "Oberyn Martell", "Ellaria Sand"]
```

```python
groupes = {}
groupes["Stark"] = ["Jon Snow", "Bran Stark", "Robb Stark", "Eddard Stark", "Sansa Stark", "Arya Stark",
                    "Catelyn Stark"]
groupes["Greyjoy"] = ["Theon Greyjoy", "Yara Greyjoy", "Euron Greyjoy"]
groupes["Lannister"] = ["Cersei Lannister", "Jaime Lannister", "Tyrion Lannister", "Tywin Lannister"]
groupes["Baratheon"] = ["Robert Baratheon", "Joffrey Baratheon", "Tommen Baratheon", "Stannis Baratheon",
                       "Renly Baratheon"]
groupes["Tyrell"] = ["Olenna Tyrell", "Margaery Tyrell", "Loras Tyrell"]
groupes["Targaryen"] = ["Daenerys Targaryen", "Viserys Targaryen"]
groupes["Martell"] = ["Oberyn Martell", "Ellaria Sand"]
groupes["Wildlings"] = ["Ygritte", "Tormund Giantsbane"]
#groupes["Night's Watch"] = ["Alliser Thorne", "Grenn", "Samwell Tarly", "Eddison Tollett"]

maisons = {}
for maison in groupes.keys():
    for perso in groupes[maison]:
        maisons[perso] = maison
```

```python
liste_lieux = {}
for ep in df['scenes']:
    for sc in ep:
        if sc['location'] not in liste_lieux.keys():
            liste_lieux[sc['location']] = 1
        else:
            liste_lieux[sc['location']] += 1
liste_lieux
```

```python
valeurs = {}
valeurs["The North"] = "The North"
valeurs["The Wall"] = "The Wall"
valeurs["North of the Wall"] = "Beyond the Wall"
valeurs["The Crownlands"] = "King's Landing"
valeurs[''] = ''
for i in ['Pentos', 'The Dothraki Sea', 'Vaes Dothrak', 'The Red Waste', 'Qarth', 'The Summer Sea', 'Astapor',
         'Yunkai', 'Meereen', 'Braavos', 'Volantis', 'Valyria']:
    valeurs[i] = "Essos"
for i in ['The Riverlands', 'The Vale', 'The Westerlands', 'Dorne', 'The Stormlands', 'The Reach',
          'The Narrow Sea', 'The Shivering Sea', 'The Sunset Sea', 'The Iron Islands']:
    valeurs[i] = "Westeros's province"
```

```python
lieux = pd.DataFrame('', index=personnages, columns=episodes)
for i, ep in enumerate(df['scenes']):
    for sc in ep:
        if all(c not in sc.keys() for c in ['warg', 'flashback']):
            characters = sc['characters']
            for ch in characters:
                if any((c in ch.keys() and ch[c]==False) for c in ['alive', 'born']):
                        continue
                if ch['name'] in lieux.index:
                    lieux.loc[ch['name']].iloc[i] = sc['location']
```

```python
def conv(ep):
    num = (int(ep[1])-1)*10 + int(ep[3:])-1
    if num >= 67:
        num -= 3
    return num
```

```python
def sequences(perso):
    liste = []
    last = ''
    for ep in perso.index:
        if valeurs[perso[ep]]!=last:
            if last=='':
                liste.append({"timeRange": [conv(ep)], "val": valeurs[perso[ep]]})
            else:
                liste[-1]['timeRange'].append(conv(ep))
                if perso[ep]!='':
                    liste.append({"timeRange": [conv(ep)], "val": valeurs[perso[ep]]})
        last = valeurs[perso[ep]]
    if last!='':
        liste[-1]['timeRange'].append(conv(ep)+1)
    return liste
```

```python
data = pd.DataFrame(index=lieux.index, columns=['data'])
data['data'] = lieux.T.apply(sequences)
```

```python
d3js = []
for maison in groupes:
    d3js.append({"group": maison, "data": []})
    for perso in groupes[maison]:
        d3js[-1]['data'].append({"label": perso, "data": data.loc[perso]['data']})
d3js.append({"group": "autres", "data": []})
for perso in data.index:
    if perso not in maisons.keys():
        d3js[-1]['data'].append({"label": perso, "data": data.loc[perso]['data']})
```

```python
index_perso = {}
for group in d3js:
    for perso in group['data']:
        index_perso[perso['label']] = len(index_perso)
with open('../data/index_perso.json', 'w') as outfile:  
    json.dump(index_perso, outfile)
```

```python
with open('../data/data.json', 'w') as outfile:  
    json.dump(d3js, outfile)
```
